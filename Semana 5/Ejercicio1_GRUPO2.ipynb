```python
 
def  uniform_cost_search(goal, start):
     
   # coste mínimo hasta
    # estado de la meta desde el inicio
    global graph,cost
    answer = []
 
    # crear una cola de prioridad
    queue = []
 
    # establecer el vector de respuesta al valor máximo
    for i in range(len(goal)):
        answer.append(10**8)
 
    # inserta el índice de inicio
    queue.append([0, start])
 
    # mapa para almacenar el nodo visitado
    visited = {}
 
    # contador
    count = 0
 
    # mientras la cola no esté vacía
    while (len(queue) > 0):
 
        # obtener elemento
        queue = sorted(queue)
        p = queue[-1]
 
        # sacar elemento 
        del queue[-1]
 
        # obtener elemento original
        p[0] *= -1
 
        # comprobar si elemento forma de la lista de meta
        if (p[1] in goal):
 
            # obtener posicion
            index = goal.index(p[1])
 
            # si se alcanza nuevo objetivo
            if (answer[index] == 10**8):
                count += 1
 
            # si el coste es menor
            if (answer[index] > p[0]):
                answer[index] = p[0]
 
            # quitar el elemento
            del queue[-1]
 
            queue = sorted(queue)
            if (count == len(goal)):
                return answer
 
        # comprueba los nodos no visitados
        # que son adyacentes al nodo actual
        if (p[1] not in visited):
            for i in range(len(graph[p[1]])):
 
                # el valor se multiplica por -1 para que
                # la menor prioridad esté en la parte superior
                queue.append( [(p[0] + cost[(p[1], graph[p[1]][i])])* -1, graph[p[1]][i]])
 
        # marcar como visitado
        visited[p[1]] = 1
 
    return answer


# función principal
if __name__ == '__main__':
     # crear el gráfico
    graph,cost = [[] for i in range(8)],{}
 
     # añadir arista
    graph[0].append(1)
    graph[0].append(3)
    graph[3].append(1)
    graph[3].append(6)
    graph[3].append(4)
    graph[1].append(6)
    graph[4].append(2)
    graph[4].append(5)
    graph[2].append(1)
    graph[5].append(2)
    graph[5].append(6)
    graph[6].append(4)
 
    #agregar costo
    cost[(0, 1)] = 2
    cost[(0, 3)] = 5
    cost[(1, 6)] = 1
    cost[(3, 1)] = 5
    cost[(3, 6)] = 6
    cost[(3, 4)] = 2
    cost[(2, 1)] = 4
    cost[(4, 2)] = 4
    cost[(4, 5)] = 3
    cost[(5, 2)] = 6
    cost[(5, 6)] = 3
    cost[(6, 4)] = 7
 
    # estado de la meta
    goal = []
 
    #establecer la meta / puede haber varios
    goal.append(6)
 
    #obtener respuesta
    answer = uniform_cost_search(goal, 0)
 
  
    print("El coste minimo del nodo 0 al 6 is = ",answer[0])
 
```
